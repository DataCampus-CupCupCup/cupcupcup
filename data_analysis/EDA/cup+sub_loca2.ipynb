```python
# pip install folium
import pandas as pd
```

```python
cup_loc = pd.read_csv('./cup_loc2.csv', encoding='EUC-KR')
```

```python
cup_loc = cup_loc[['loc', 'lat', 'long']]
cup_loc
```

```python
import folium
# map_osm = folium.Map(location=[37.483559,127.032618], zoom_start=12) #지도 초기 위치 : 서초구청

map_osm =folium.Map(location=[cup_loc['lat'].mean(), cup_loc['long'].mean()] ,zoom_start=13)
```

## 서리풀컵 위치 표시

```python
# for item in cup_loc.index:
#     lat = cup_loc.loc[item, 'lat']
#     long = cup_loc.loc[item, 'long']
#     folium.Marker([lat, long], popup = cup_loc.loc[item, 'loc'],
#                   color = 'blue', fill = True).add_to(map_osm)
# map_osm

for n1,n2,n3 in cup_loc.values:
    folium.Marker(location= [n2, n3], popup = n1).add_to(map_osm)
map_osm
```

## 지하철역 위치 표시

```python
sub_loc = pd.read_csv('./subway_loc.csv', encoding='EUC-KR')
sub_loc = sub_loc[sub_loc['loc1'] == '서초구'][['station_name', 'lat', 'long']].drop_duplicates(['lat'], keep = 'first') # 중복행 제거
sub_loc
```

```python
# for item in sub_loc.index:
#     lat = sub_loc.loc[item, 'lat']
#     long = sub_loc.loc[item, 'long']
#     folium.CircleMarker([lat, long], radius=30,
#                   color = 'red', fill = True).add_to(map_osm)
# map_osm


for n1,n2,n3 in sub_loc.values:
    folium.CircleMarker(location= [n2, n3], popup = n1, radius=20, color = 'red', fill = True).add_to(map_osm)
map_osm
```
